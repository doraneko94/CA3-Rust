```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
x = np.linspace(-80, 40, 200)
```

```python
E_Na = 55
E_K = -90
E_L = -65
```

```python
phi = 5
```

```python
g_Na = 35
g_K = 9
g_L = 0.1
```

```python
def alpha_m(v):
    return -0.1 * (v + 35) / (np.exp(-0.1 * (v + 35)) - 1)
plt.plot(x, alpha_m(x))
```

```python
def beta_m(v):
    return 4 * np.exp(-(v + 60) / 18)
plt.plot(x, beta_m(x))
```

```python
def m_inf(v):
    return alpha_m(v) / (alpha_m(v) + beta_m(v))
plt.plot(x, m_inf(x))
```

```python
def alpha_h(v):
    return 0.07 * np.exp(-(v + 58) / 20)
plt.plot(x, alpha_h(x))
```

```python
def beta_h(v):
    return 1 / (np.exp(-0.1 * (v + 28)) + 1)
plt.plot(x, beta_h(x))
```

```python
def h_inf(v):
    return alpha_h(v) / (alpha_h(v) + beta_h(v))
plt.plot(x, h_inf(x))
```

```python
def i_na(v):
    return g_Na * m_inf(v)**3 * h_inf(v) * (v - E_Na)
plt.plot(x, i_na(x))
```

```python
def alpha_n(v):
    return -0.01 * (v + 34) / (np.exp(-0.1 * (v + 34)) - 1)
plt.plot(x, alpha_n(x))
```

```python
def beta_n(v):
    return 0.125 * np.exp(-(v + 44) / 80)
plt.plot(x, beta_n(x))
```

```python
def n_inf(v):
    return alpha_n(v) / (alpha_n(v) + beta_n(v))
plt.plot(x, n_inf(x))
```

```python
def i_k(v):
    return g_K * n_inf(v)**4 * (v - E_K)
plt.plot(x, i_k(x))
```

```python
def i(v):
    return - i_na(v) - i_k(v) - g_L * (v - E_L)
plt.plot(x, i(x))
plt.ylim([-1, 10])
```
