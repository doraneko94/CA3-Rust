```python
df = pd.read_csv("fig1b.csv", index_col=0)
df.columns
```

```python
df["g_c=2.1|clamp=60"].plot()
```

```python
gc0018_max = df.max(axis=0)[0:10].values
gc0060_max = df.max(axis=0)[10:20].values
gc0210_max = df.max(axis=0)[20:30].values
gc0360_max = df.max(axis=0)[30:40].values
gc0900_max = df.max(axis=0)[40:50].values
gc1500_max = df.max(axis=0)[50:60].values
```

```python
gc0018_dss = df.iloc[len(df)-1][0:10].values
gc0060_dss = df.iloc[len(df)-1][10:20].values
gc0210_dss = df.iloc[len(df)-1][20:30].values
gc0360_dss = df.iloc[len(df)-1][30:40].values
gc0900_dss = df.iloc[len(df)-1][40:50].values
gc1500_dss = df.iloc[len(df)-1][50:60].values
```

```python
clamp_hold = np.array([i*10 + 10 for i in range(10)])
```

```python
gc0018 = (gc0018_max - gc0018_dss) / clamp_hold
gc0060 = (gc0060_max - gc0060_dss) / clamp_hold
gc0210 = (gc0210_max - gc0210_dss) / clamp_hold
gc0360 = (gc0360_max - gc0360_dss) / clamp_hold
gc0900 = (gc0900_max - gc0900_dss) / clamp_hold
gc1500 = (gc1500_max - gc1500_dss) / clamp_hold
```

```python
x = np.array([i*10 for i in range(10)])
```

```python
plt.plot(x, gc0018)
plt.plot(x, gc0060)
plt.plot(x, gc0210)
plt.plot(x, gc0360)
plt.plot(x, gc0900)
plt.plot(x, gc1500)
plt.xlabel("V_clamp (mV)")
plt.ylabel("Transient Attenuation Factor")
plt.xlim([0.0, 100.0])
plt.ylim([0.0, 2.0])
plt.title("fig.1b")
plt.legend(["g_c=0.18", "g_c=0.6", "g_c=2.1", "g_c=3.6", "g_c=9.0", "g_c=15.0"])
plt.savefig("fig1b.png")
```
