```python
df = pd.read_csv("fig3.csv", index_col=0)
df.columns
```

```python
ax1 = df[["V_s(mV)", "V_d(mV)"]].plot()
ax1.set_ylim([-10,90])
ax1.set_ylabel("Membrane Potential (mV)")
ax2 = ax1.twinx()
ax2.set_ylim([0, 800])
ax2.set_ylabel("Ca")
ax2.plot(df.index, df["Ca"], color='#2ca02c', label="Ca")
h1, _ = ax1.get_legend_handles_labels()
h2, l2 = ax2.get_legend_handles_labels()
ax1.legend(h1+h2, ["V_s", "V_d"]+l2, loc='upper right')
plt.xlim([460, 510])
plt.title("fig.3")
plt.savefig("fig3.png")
```
