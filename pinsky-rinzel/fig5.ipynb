```python
df = pd.read_csv("fig5.csv", index_col=0)
df.columns
```

```python
def hz(arr):
    ans = 0
    flg = False
    for v in arr:
        if not flg and v > 70:
            ans += 1
        if v > 70:
            flg = True
        else:
            flg = False
    return ans
```

```python
somas = ['soma-0.175', 'soma-0.165', 'soma-0.125', 'soma0', 'soma0.25',
       'soma0.5', 'soma0.75', 'soma1', 'soma1.25', 'soma1.5', 'soma1.75',
       'soma2', 'soma2.25', 'soma2.5', 'soma2.75', 'soma3']
dends = ['dend0.25',
       'dend0.5', 'dend0.75', 'dend1', 'dend1.25', 'dend1.5', 'dend1.75',
       'dend2', 'dend2.25', 'dend2.5', 'dend2.75', 'dend3']
gc21is = ['g_c21i_s-0.25',
       'g_c21i_s-0.125', 'g_c21i_s0', 'g_c21i_s0.25', 'g_c21i_s0.5',
       'g_c21i_s0.75', 'g_c21i_s1', 'g_c21i_s1.25', 'g_c21i_s1.5',
       'g_c21i_s1.75', 'g_c21i_s2', 'g_c21i_s2.25', 'g_c21i_s2.5',
       'g_c21i_s2.75', 'g_c21i_s3']
gc21id = ['g_c21i_d-0.25', 'g_c21i_d-0.125',
       'g_c21i_d0', 'g_c21i_d0.25', 'g_c21i_d0.5', 'g_c21i_d0.75', 'g_c21i_d1',
       'g_c21i_d1.25', 'g_c21i_d1.5', 'g_c21i_d1.75', 'g_c21i_d2',
       'g_c21i_d2.25', 'g_c21i_d2.5', 'g_c21i_d2.75', 'g_c21i_d3']
gcinf = ['g_cinf-0.25', 'g_cinf-0.125', 'g_cinf0', 'g_cinf0.25', 'g_cinf0.5',
       'g_cinf0.75', 'g_cinf1', 'g_cinf1.25', 'g_cinf1.5', 'g_cinf1.75',
       'g_cinf2', 'g_cinf2.25', 'g_cinf2.5', 'g_cinf2.75', 'g_cinf3']
```

```python
y_somas = [hz(df[n]) / 10 for n in somas]
y_dends = [hz(df[n]) / 10 for n in dends]
y_gc21is = [hz(df[n]) / 10 for n in gc21is]
y_gc21id = [hz(df[n]) / 10 for n in gc21id]
y_gcinf = [hz(df[n]) / 10 for n in gcinf]
```

```python
x_somas = [-0.175, -0.165, -0.125, 0, 0.25, 0.5, 0.75, 1, 1.25, 1.5, 1.75, 2, 2.25, 2.5, 2.75, 3]
x_dends = [0.25, 0.5, 0.75, 1, 1.25, 1.5, 1.75, 2, 2.25, 2.5, 2.75, 3]
x_neurons = [-0.25, -0.125, 0.0, 0.25, 0.5, 0.75, 1, 1.25, 1.5, 1.75, 2, 2.25, 2.5, 2.75, 3]
```

```python
plt.plot(x_somas, y_somas, marker="^", label="Soma alone")
plt.plot(x_dends, y_dends, marker="^", label="Dendrite alone")
plt.plot(x_neurons, y_gc21is, marker="o", label="g_c=2.1 (I_s)")
plt.plot(x_neurons, y_gc21id, marker="o", label="g_c=2.1 (I_d)")
plt.plot(x_neurons, y_gcinf, marker="^", label="g_c=∞")
plt.yscale("log")
plt.xlim([-0.6, 3.1])
plt.ylim([0.2, 100])
plt.legend()
plt.xlabel("I (uA/cm2)")
plt.ylabel("Frequency (Hz)")
plt.title("Fig.5")
plt.savefig("fig5.png")
```
