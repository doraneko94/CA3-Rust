```python
df = pd.read_csv("fig2.csv", index_col=0)
df.columns
```

```python
df["A|V_s(mV)"].plot()
plt.xlim([0, 1600])
plt.ylim([-10, 100])
plt.legend(["V_s(mV)"])
plt.ylabel("V_s(mV)")
plt.title("fig.2A left")
plt.savefig("fig2AL.png")
```

```python
ax1 = df["A|Ca"].plot(label="Ca")
ax1.set_ylim([0,500])
ax1.set_ylabel("Ca")
ax2 = ax1.twinx()
ax2.set_ylim([0, 0.7])
ax2.set_ylabel("q")
ax2.plot(df.index, df["A|q"], color='#ff7f0e', label="q")
h1, l1 = ax1.get_legend_handles_labels()
h2, l2 = ax2.get_legend_handles_labels()
ax1.legend(h1+h2, l1+l2, loc='upper right')
plt.xlim([0, 1600])
plt.title("fig.2A right")
plt.savefig("fig2AR.png")
```
