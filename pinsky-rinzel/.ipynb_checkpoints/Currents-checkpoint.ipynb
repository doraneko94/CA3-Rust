```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
x = np.linspace(-20, 100, 200)
```

```python
V_Na = 120
V_Ca = 140
V_K = -15
```

```python
def alpha_m(v):
    return 0.32 * (13.1 - v) / (np.exp((13.1 - v) / 4) - 1)
plt.plot(x, alpha_m(x))
```

```python
def beta_m(v):
    return 0.28 * (v - 40.1) / (np.exp((v - 40.1) / 5) - 1)
plt.plot(x, beta_m(x))
```

```python
def m_inf(v):
    return alpha_m(v) / (alpha_m(v) + beta_m(v))
plt.plot(x, m_inf(x))
```

```python
m_inf(-4.6)
```

```python
def alpha_h(v):
    return 0.128 * np.exp((17 - v) / 18)
plt.plot(x, alpha_h(x))
```

```python
def beta_h(v):
    return 4 / (1 + np.exp((40 - v) / 5))
plt.plot(x, beta_h(x))
```

```python
def h_inf(v):
    return alpha_h(v) / (alpha_h(v) + beta_h(v))
plt.plot(x, h_inf(x))
```

```python
h_inf(-4.6)
```

```python
def tau_h(v):
    return 1 / (alpha_h(v) + beta_h(v))
plt.plot(x, tau_h(x))
```

```python
def i_na(v):
    return m_inf(v)**2 * h_inf(v) * (v - V_Na)
plt.plot(x, i_na(x))
```

```python
def alpha_n(v):
    return 0.016 * (35.1 - v) / (np.exp((35.1 - v) / 5) - 1)
plt.plot(x, alpha_n(x))
```

```python
def beta_n(v):
    return 0.25 * np.exp(0.5 - 0.025 * v)
plt.plot(x, beta_n(x))
```

```python
def n_inf(v):
    return alpha_n(v) / (alpha_n(v) + beta_n(v))
plt.plot(x, n_inf(x))
```

```python
n_inf(-4.6)
```

```python
def tau_n(v):
    return 1 / (alpha_n(v) + beta_n(v))
plt.plot(x, tau_n(x))
```

```python
def i_k_dr(v):
    return n_inf(v) * (v - V_K)
plt.plot(x, i_k_dr(x))
```

```python
def alpha_s(v):
    return 1.6 / (1 + np.exp(-0.072 * (v - 65)))
plt.plot(x, alpha_s(x))
```

```python
def beta_s(v):
    return 0.02 * (v - 51.1) / (np.exp((v - 51.1) / 5) - 1)
plt.plot(x, beta_s(x))
```

```python
def s_inf(v):
    return alpha_s(v) / (alpha_s(v) + beta_s(v))
plt.plot(x, s_inf(x))
```

```python
s_inf(-4.5)
```

```python
def tau_s(v):
    return 1 / (alpha_s(v) + beta_s(v))
plt.plot(x, tau_s(x))
```

```python
def i_ca(v):
    return s_inf(v)**2 * (v - V_Ca)
plt.plot(x, i_ca(x))
```

```python
x_ca = np.linspace(0.0, 800.0, 800)
```

```python
def kai(ca):
    if type(ca) == np.ndarray:
        ret = ca / 250
        ret[ret > 1] = 1
        return ret
    else:
        return min(ca / 250, 1)
plt.plot(x_ca, kai(x_ca))
```

```python
kai(100)
```

```python
def alpha_c(v):
    if type(v) == np.ndarray:
        ret_l = np.exp((v - 10) / 11 - (v - 6.5) / 27) / 18.975
        ret_u = 2 * np.exp((6.5 - v) / 27)
        ret_l[v > 50] = ret_u[v > 50]
        return ret_l
    else:
        if v <= 50:
            return np.exp((v - 10) / 11 - (v - 6.5) / 27) / 18.975
        else:
            return 2 * np.exp((6.5 - v) / 27)
plt.plot(x, alpha_c(x))
```

```python
def beta_c(v):
    if type(v) == np.ndarray:
        ret_l = 2 * np.exp((6.5 - v) / 27) - alpha_c(v)
        ret_u = np.zeros_like(v)
        ret_l[v > 50] = ret_u[v > 50]
        return ret_l
    else:
        if v <= 50:
            return 2 * np.exp((6.5 - v) / 27) - alpha_c(v)
        else:
            return 0
plt.plot(x, beta_c(x))
```

```python
def c_inf(v):
    return alpha_c(v) / (alpha_c(v) + beta_c(v))
plt.plot(x, c_inf(x))
```

```python
c_inf(-4.5)
```

```python
def tau_c(v):
    return 1 / (alpha_c(v) + beta_c(v))
plt.plot(x, tau_c(x))
```

```python
def i_k_c(v, ca):
    return c_inf(v) * kai(ca) * (v - V_K)
plt.plot(x, i_k_c(x, 0.2))
```

```python
from mpl_toolkits.mplot3d import Axes3D

X1, X2 = np.meshgrid(x, x_ca)
X = np.c_[np.ravel(X1), np.ravel(X2)]
Z = np.array([i_k_c(x, ca) for ca in x_ca])
Z.reshape(X1.shape)
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
surf = ax.plot_surface(X1, X2, Z, cmap='bwr', linewidth=0)
fig.colorbar(surf)
```

```python
def alpha_q(ca):
    if type(ca) == np.ndarray:
        ret = ca * 0.2 * 1e-4
        ret[ret > 0.01] = 0.01
        return ret
    else:
        return min(ca * 0.2 * 1e-4, 0.01)
plt.plot(x_ca, alpha_q(x_ca))
```

```python
def q_inf(ca):
    return alpha_q(ca) / (alpha_q(ca) + 0.001)
plt.plot(x_ca, q_inf(x_ca))
```

```python
q_inf(0.2)
```

```python
def tau_q(ca):
    return 1 / (alpha_q(ca) + 0.001)
plt.plot(x_ca, tau_q(x_ca))
```

```python
def i_k_ahp(v, ca):
    return q_inf(ca) * (v - V_K)
plt.plot(x, i_k_ahp(x, 0.2))
```

```python
X1, X2 = np.meshgrid(x, x_ca)
X = np.c_[np.ravel(X1), np.ravel(X2)]
Z = np.array([i_k_ahp(x, ca) for ca in x_ca])
Z.reshape(X1.shape)
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
surf = ax.plot_surface(X1, X2, Z, cmap='bwr', linewidth=0)
fig.colorbar(surf)
```
